# Caminhadas Aleatórias

## Imports

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (12,10)

from tqdm.notebook import tqdm
```

## Funções

```python
def euclid_dist(x:np.array,y:np.array):
    return np.linalg.norm(x-y)
```

```python
def perform_1D_RW(num_steps:int,step_size_interval:tuple,seed:int):

    np.random.seed(seed)
    coord = np.zeros([num_steps+1])
    traveled_dist = 0

    for curr_step in range(1,num_steps+1):
        
        coord[curr_step] = np.random.uniform(*step_size_interval)
        traveled_dist += np.abs(coord[curr_step] - coord[curr_step-1])
        
    trajectory = np.cumsum(coord)
    
    return trajectory,traveled_dist


def perform_2D_RW(num_steps:int,step_size_interval:tuple,seed:int):

    np.random.seed(seed)
    coord = np.zeros([num_steps+1,2])
    traveled_dist = 0

    for curr_step in range(1,num_steps+1):
        theta = np.random.uniform(0,2*np.pi)
        step_size_x = np.random.uniform(*step_size_interval)
        step_size_y = np.random.uniform(*step_size_interval) 
        
        coord[curr_step,0] = step_size_x
        coord[curr_step,1] = step_size_y 
        
        traveled_dist += euclid_dist(coord[curr_step,:],coord[curr_step-1,:])
        
    trajectories = np.cumsum(coord,axis=0)
    
    return trajectories,traveled_dist
```

## Experimento 1

### Letra a)

```python
num_steps = 10000
step_size_interval = (-0.5,0.5)
```

```python
plt.xlabel("Passo t")
plt.ylabel("Posição x")
plt.title(f"Caminhada aleatória 1D para N = {num_steps}")

for seed in tqdm(np.arange(10)):
    trajectory,traveled_dist = perform_1D_RW(num_steps,step_size_interval,seed)
    
    plt.plot(np.arange(len(trajectory)),trajectory,label=f"Dist = {traveled_dist:.2f}")
    

plt.legend()
plt.show()
```

### Letra b)

```python
plt.xlabel("Posição x")
plt.ylabel("Posição y")
plt.title(f"Caminhada aleatória 2D")

for num_steps in tqdm([10,1000,100000]):
    for seed in np.arange(3):
        trajectories,traveled_dist = perform_2D_RW(num_steps,step_size_interval,seed)

        plt.plot(trajectories[:,0],trajectories[:,1],label=f"N={num_steps},Dist = {traveled_dist:.2f}")


plt.legend()
plt.show()
```

* É possível observar que, de fato, a distância aparenta aumentar 10 vezes quando se aumenta o número de passos em 100 vezes (de 1000 para 100000).

## Experimento 2

```python
def compute_RW_final_points(rw_step,num_steps:int,num_walks:int,step_size_interval:tuple,seeds:list=None):
    final_points = []
    if seeds == None:
        seeds = np.arange(num_walks)
    
    for i in range(num_walks):
        trajectories,_ = rw_step(num_steps,step_size_interval,seeds[i])
        final_points.append(trajectories[-1])
        
    axis = None
    if len(final_points[0].shape) == 1:
        axis = 1
    return np.concatenate([final_points],axis=axis)
```

```python
num_walks = 10000
step_size_interval = (-0.5,0.5)
```

```python
plt.xlabel("Posição x")
plt.ylabel("Posição y")
plt.title(f"Posição final de W = {num_walks} caminhadas aleatórias 2D")

for num_steps in tqdm([10,1]):
    final_points = compute_RW_final_points(perform_2D_RW,num_steps,num_walks,step_size_interval)
    plt.scatter(final_points[:,0],final_points[:,1],label=f"N={num_steps}")

plt.legend()
plt.show()
```

### Letra c)

```python
num_walks = 10000
step_size_interval = (-0.5,0.5)
```

```python
uniform_std =np.sqrt(1/12 * np.power((step_size_interval[1]-step_size_interval[0]),2))# desvio padrão de distribuição uniforme

for num_steps in tqdm([1,2,3,5]):
    final_points = compute_RW_final_points(perform_1D_RW,num_steps,num_walks,step_size_interval)
    
    #Estimativa gaussiana
    sigma = np.sqrt(num_steps) * uniform_std
    
    xmin = -3*sigma
    xmax = 3*sigma
    x = np.linspace(xmin,xmax)
        
    gaussian_est = 1/(np.sqrt(2*np.pi*sigma)) * np.exp(-(np.power(x,2)/2*np.power(sigma,2)))
    
    
    plt.ylabel("# pontos")
    plt.xlabel("Posição x")
    plt.title(f"Posição final de W = {num_walks} caminhadas aleatórias 1D com N = {num_steps} passos")
    
    plt.xlim(xmin,xmax)
    
    plt.hist(final_points,label=f"N={num_steps}",bins=50,density=True)
    plt.plot(x,gaussian_est,'r')
    plt.show()

```

* A distribuição gaussiana se torna uma boa estimativa de forma relativamente rápida. Podemos ver que para N = 5, a estimativa gaussiana ja é razoável.
